```python
import numpy as np
import matplotlib.pyplot as plt

n_samples = 100000
X1 = np.random.uniform(0, 1, n_samples)
X2 = np.random.uniform(0, 1, n_samples)
Z = X1 + X2

plt.figure(figsize=(8, 5))
plt.hist(Z, bins=100, density=True, alpha=0.7, label='Empirical PDF', color='skyblue')

z_vals = np.linspace(0, 2, 500)
pdf_vals = np.where(z_vals < 1, z_vals, 2 - z_vals)
plt.plot(z_vals, pdf_vals, label='Theoretical PDF', color='red', linewidth=2)

plt.title('Distribution of Z = X1 + X2, where X1, X2 ~ Uniform(0,1)')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n_samples = 100000
X1 = np.random.normal(0, 1, n_samples)
X2 = np.random.normal(0, 1, n_samples)
Z = X1 + X2

plt.figure(figsize=(8, 5))
plt.hist(Z, bins=100, density=True, alpha=0.6, label='Empirical PDF', color='lightgreen')

z_vals = np.linspace(-6, 6, 1000)
pdf_vals = norm.pdf(z_vals, loc=0, scale=np.sqrt(2))
plt.plot(z_vals, pdf_vals, 'r', lw=2, label='Theoretical PDF (N(0, 2))')

plt.title('Distribution of Z = X1 + X2, where X1, X2 ~ N(0,1)')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_samples = 100000
X1 = np.random.uniform(0, 1, n_samples)
X2 = np.random.uniform(0, 2, n_samples)
Z = X1 + X2

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(Z, bins=100, density=True, alpha=0.6, color='orange', label='Empirical PDF')

# Theoretical PDF (trapezoidal)
z_vals = np.linspace(0, 3, 1000)
pdf_vals = np.piecewise(
    z_vals,
    [z_vals < 1, (z_vals >= 1) & (z_vals < 2), z_vals >= 2],
    [lambda z: z / 2, 0.5, lambda z: (3 - z) / 2]
)
plt.plot(z_vals, pdf_vals, 'r', linewidth=2, label='Theoretical PDF')

plt.title('Distribution of Z = X1 + X2 (X1~U(0,1), X2~U(0,2))')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```
