```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

file_path = r"C:\Users\USER\Downloads\people_25_dataset.csv"
df = pd.read_csv(file_path)

heights = df['Height (cm)']
mean_h = np.mean(heights)
std_h = np.std(heights)
x_h = np.linspace(min(heights), max(heights), 100)
pdf_h = norm.pdf(x_h, mean_h, std_h)
cdf_h = norm.cdf(x_h, mean_h, std_h)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(heights, bins=10, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.plot(x_h, pdf_h, 'r-')
plt.title('PDF of Height (cm)')
plt.subplot(1, 2, 2)
plt.plot(x_h, cdf_h, 'g-')
plt.title('CDF of Height (cm)')
plt.tight_layout()
plt.show()

weights = df['Weight (kg)']
mean_w = np.mean(weights)
std_w = np.std(weights)
x_w = np.linspace(min(weights), max(weights), 100)
pdf_w = norm.pdf(x_w, mean_w, std_w)
cdf_w = norm.cdf(x_w, mean_w, std_w)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(weights, bins=10, density=True, alpha=0.6, color='orange', edgecolor='black')
plt.plot(x_w, pdf_w, 'r-')
plt.title('PDF of Weight (kg)')
plt.subplot(1, 2, 2)
plt.plot(x_w, cdf_w, 'g-')
plt.title('CDF of Weight (kg)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.style.use('seaborn-v0_8-whitegrid')

file_path =  r"C:\Users\USER\Downloads\people_25_dataset.csv"
df = pd.read_csv(file_path)

heights = df['Height (cm)'].dropna().to_numpy()
weights = df['Weight (kg)'].dropna().to_numpy()

# Heights: Normal PDF and CDF
data = heights
mean = np.mean(data)
std = np.std(data)
x = np.linspace(min(data), max(data), 200)
pdf = norm.pdf(x, mean, std)
cdf = norm.cdf(x, mean, std)

plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'b')
plt.title('PMF: Probability Density Function (Height)')
plt.xlabel('Height (cm)')
plt.ylabel('P(X = x)')
plt.grid(True, linestyle='--', linewidth=0.7)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'orange')
plt.title('CDF: Cumulative Distribution Function (Height)')
plt.xlabel('Height (cm)')
plt.ylabel('P(X  x)')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()

# Weights: Normal PDF and CDF
data = weights
mean = np.mean(data)
std = np.std(data)
x = np.linspace(min(data), max(data), 200)
pdf = norm.pdf(x, mean, std)
cdf = norm.cdf(x, mean, std)

plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'b')
plt.title('PMF: Probability Density Function (Weight)')
plt.xlabel('Weight (kg)')
plt.ylabel('P(X = x)')
plt.grid(True, linestyle='--', linewidth=0.7)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'orange')
plt.title('CDF: Cumulative Distribution Function (Weight)')
plt.xlabel('Weight (kg)')
plt.ylabel('P(X  x)')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```
