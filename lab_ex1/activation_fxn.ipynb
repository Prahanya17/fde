```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
dy = sigmoid_derivative(x)

plt.plot(x, y, label='Sigmoid', color='blue')
plt.plot(x, dy, label="Sigmoid Derivative", color='cyan', linestyle='--')
plt.title('Sigmoid and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

x = np.linspace(-10, 10, 1000)
y = tanh(x)
dy = tanh_derivative(x)

plt.plot(x, y, label='Tanh', color='green')
plt.plot(x, dy, label="Tanh Derivative", color='lime', linestyle='--')
plt.title('Tanh and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

x = np.linspace(-10, 10, 1000)
y = relu(x)
dy = relu_derivative(x)

plt.plot(x, y, label='ReLU', color='red')
plt.plot(x, dy, label="ReLU Derivative", color='orange', linestyle='--')
plt.title('ReLU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

x = np.linspace(-10, 10, 1000)
y = leaky_relu(x)
dy = leaky_relu_derivative(x)

plt.plot(x, y, label='Leaky ReLU', color='purple')
plt.plot(x, dy, label="Leaky ReLU Derivative", color='magenta', linestyle='--')
plt.title('Leaky ReLU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = np.exp(-x)
dy = -np.exp(-x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = e^(-x)', color='orange')
plt.plot(x, dy, label="y' = -e^(-x)", color='blue', linestyle='--')
plt.title('y = e^(-x) and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-np.abs(x))

def f_derivative(x):
    return -np.sign(x) * np.exp(-np.abs(x))

x = np.linspace(-10, 10, 1000)
y = f(x)
dy = f_derivative(x)

plt.plot(x, y, label='y = e^(-|x|)', color='teal')
plt.plot(x, dy, label="Derivative", color='orange', linestyle='--')
plt.title('y = e^(-|x|) and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)

def f_derivative(x):
    return -2 * x * np.exp(-x**2)

x = np.linspace(-3, 3, 1000)
y = f(x)
dy = f_derivative(x)

plt.plot(x, y, label='y = e^(-x²)', color='darkblue')
plt.plot(x, dy, label="Derivative", color='crimson', linestyle='--')
plt.title('y = e^(-x²) and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```
